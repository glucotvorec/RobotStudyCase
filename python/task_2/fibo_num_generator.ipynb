```python
from itertools import islice
"""Функция для генерации чисел Фибоначчи"""

def fib():
    # Начальные значения двух первых элементов последовательности
    prev, curr = 0, 1
    
    # Бесконечный цикл для генерации всех чисел Фибоначчи
    while True:
        yield curr
        prev, curr = curr, prev + curr

def get_fib_numbers(max_value=100):
    """ Функция для получения списка чисел Фибоначчи до заданного предела"""
    
    # Создаем генератор чисел Фибоначчи
    f = fib()
    
    # Используем функцию islice для получения нужного количества чисел
    return list(islice(f, max_value))


# Получение списка чисел Фибоначчи
fib_numbers = get_fib_numbers() 

# Проверка длины полученного списка
assert len(fib_numbers) == 100

# Выводим список чисел Фибоначчи
print(fib_numbers)
```
